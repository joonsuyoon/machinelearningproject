인구밀도 데이터 전처리

```python
import pandas as pd

df = pd.read_csv('./datasets/인구밀도_2019~2020.csv')

# 첫 행 드랍
df.drop(0, inplace=True)

# 동별(3)이 소계인 row를 삭제
df.drop(df[df['동별(3)'] == '소계'].index, inplace=True)

# 동별(1) 열을 드롭
df.drop('동별(1)', axis=1, inplace=True)

# 동별(2)를 구, 동별(3)를 동으로 이름 바꿈
df.rename(columns={'동별(2)': '구', '동별(3)': '동'}, inplace=True)

# 한 row에 중복되는 행 2019, 2019.1, 2019.2, 2020, 2020.1, 2020.2을 인구 2019, 인구밀도 두 row로 분리
id_cols = ['구', '동']

cols_2019 = [col for col in df.columns if '2019' in col]
df_2019 = df[id_cols + cols_2019].copy()
df_2019['연도'] = '2019'
df_2019.rename(columns={'2019': '인구', '2019.1': '면적', '2019.2': '인구밀도'}, inplace=True)

cols_2020 = [col for col in df.columns if '2020' in col]
df_2020 = df[id_cols + cols_2020].copy()
df_2020['연도'] = '2020'
df_2020.rename(columns={'2020': '인구', '2020.1': '면적', '2020.2': '인구밀도'}, inplace=True)

인구밀도 = pd.concat([df_2019, df_2020])
인구밀도.to_csv('./refined_datasets/인구밀도.csv', index=False)

인구밀도
```

```python
import pandas as pd

df = pd.read_csv('./datasets/서울특별시 강서구_행정동별 쓰레기 배출량_20201231.csv')

df.rename(columns={'행정동': '동'}, inplace=True)

id_cols = ['동']

cols_2019 = [col for col in df.columns if '2019' in col]
df_2019 = df[id_cols + cols_2019].copy()
df_2019['쓰레기배출량'] = df_2019[cols_2019].sum(axis=1)
df_2019 = df_2019[['동', '쓰레기배출량']]
df_2019['연도'] = '2019'
# kg에서 t로
df_2019['쓰레기배출량'] = df_2019['쓰레기배출량'] / 1000

cols_2020 = [col for col in df.columns if '2020' in col]
df_2020 = df[id_cols + cols_2020].copy()
df_2020['쓰레기배출량'] = df_2020[cols_2020].sum(axis=1)
df_2020 = df_2020[['동', '쓰레기배출량']]
df_2020['연도'] = '2020'
# kg에서 t로
df_2020['쓰레기배출량'] = df_2020['쓰레기배출량'] / 1000

강서구_쓰레기배출량 = pd.concat([df_2019, df_2020])
강서구_쓰레기배출량.to_csv('./refined_datasets/강서구_쓰레기배출량.csv', index=False)

강서구_쓰레기배출량
```

```python
import pandas as pd

df = pd.read_csv('./datasets/서울시 단란주점영업 인허가 정보.csv')
codes = pd.read_excel('./datasets/자치단체코드목록.xlsx')

df = df[['개방자치단체코드', '상세영업상태명', '폐업일자', '소재지면적', '좌표정보(X)', '좌표정보(Y)']]

# 폐업일자가 2019년 1월 1일 이후거나 NaN인 행만 남김
df = df[(df['폐업일자'].isna()) | (df['폐업일자'] >= '20190101')]

df = df.merge(codes, left_on='개방자치단체코드', right_on='시군구(자치단체)코드', how='left')
df = df[['시군구명', '상세영업상태명', '폐업일자', '소재지면적', '좌표정보(X)', '좌표정보(Y)']]

df.rename(columns={'시군구명': '구', '상세영업상태명': '상태', '소재지면적': '면적', '좌표정보(X)': 'X좌표', '좌표정보(Y)': 'Y좌표'}, inplace=True)
df.dropna(subset=['구'], inplace=True)

# 좌표 변경
from pyproj import Proj, transform
proj_wgs84 = Proj(init='epsg:4326')  # WGS84 좌표계
proj_bessel = Proj(init='epsg:2097')  # 2097 좌표계

converted_x, converted_y = transform(proj_bessel, proj_wgs84, df['X좌표'].values, df['Y좌표'].values)
df['X좌표'] = converted_x
df['Y좌표'] = converted_y
df.rename(columns={'X좌표': '경도', 'Y좌표': '위도'}, inplace=True)
df.dropna(subset=['경도', '위도'], inplace=True)

df.to_csv('./refined_datasets/단란주점영업_인허가정보.csv', index=False)
df
```
